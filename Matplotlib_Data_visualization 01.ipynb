```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#Matplotlib 基本操作
#假設要劃一個sin波 從0到180 
x = np.arange(0,180)
y = np.sin(x*np.pi/ 180.0)
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(0,180)
y = np.sin(x*np.pi/ 180.0)
plt.plot(x,y)
#設定圖的範圍
plt.xlim(-30,360)
plt.ylim(-1,2)
#寫入x,y軸的label and Title
plt.xlabel("this is X-axis")
plt.ylabel("this is Y's")
plt.title("Title of the graph")

plt.show
```

```python
#Scatter Plots
X = np.random.normal(0,1,500)  #normal distribution (mean,stddev,amount)
Y = np.random.normal(0,1,500)
plt.scatter(X, Y)
plt.title("Scatter plot")

n = 128
X = np.random.normal(0,1,n)  
Y = np.random.normal(0,1,n)
plt.scatter(X,Y)
plt.show()
```

```python
#Colour and shape setting
t = np.arange(1,10, 0.4)
print("t:",str(t))
plt.plot(t,t,'r--',t,t**2,'bs',t,t**3,'g^')
```

```python
t = np.arange(0,10,0.4)
plt.plot(t,t,'r--',label="y=x")
plt.plot(t,t**2,'bs',label="y=x^2")
plt.plot(t,t**3,'g^',label="y=x^3")
plt.legend() #圖例
```

```python
x =np.arange(0,10, 0.4)
y =np.arange(0,5, 0.2)
plt.bar(x,y)
```

```python
#進階_單線圖
ts =pd.Series(np.random.randn(200),index=pd.date_range('1/1/2021',periods=200))
ts = ts.cumsum()  #.cumsum() :Return cumulative sum over a DataFrame or Series axis.
print(ts)
ts.plot()
```

```python
#多個折線圖同時劃出
df = pd.DataFrame(np.random.randn(200,3),index=pd.date_range('01/01/2021',periods=200),columns=['A','B','C'])
df = df.cumsum()
print(df)
df.plot()
```

```python
#Pie Chart
df = pd.Series(np.random.rand(4),index=["A","B","C","D"],name="Title")
print(df)
df.plot.pie()
```

```python
#Bar Chart
df = pd.DataFrame(np.random.rand(10,2),columns=["A","B"])
print(df)
df.plot.bar()
```

```python
df.plot.bar(stacked=True)
```

```python
#Boxplot
df=pd.DataFrame(np.random.rand(10,4),columns=["A","B","D","E"])
print(df)
df.boxplot()
```
